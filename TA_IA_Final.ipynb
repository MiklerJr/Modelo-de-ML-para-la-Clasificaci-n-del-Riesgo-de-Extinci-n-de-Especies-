# üìò **Descripci√≥n del Dataset de Especies Terrestres en Riesgo de Extinci√≥n**

En esta experiencia, trabajaremos con un dataset que contiene datos ecol√≥gicos, biol√≥gicos y geogr√°ficos sobre diversas especies terrestres. El objetivo es **clasificar el riesgo de extinci√≥n** de estas especies en categor√≠as definidas por la **UICN** (Uni√≥n Internacional para la Conservaci√≥n de la Naturaleza), utilizando una variedad de rasgos morfol√≥gicos y ambientales.

En este an√°lisis, las especies se clasifican en **categor√≠as de amenaza** que incluyen:  
- **LC**: Preocupaci√≥n Menor  
- **VU**: Vulnerable  
- **EN**: En Peligro  
- **CR**: En Peligro Cr√≠tico

### **Transformaci√≥n de la columna `Status`**:

La columna **`Status`** originalmente conten√≠a las siguientes categor√≠as: **LC**, **NT**, **VU**, **EN**, **CR**, **EW**, **EX**. Para simplificar la clasificaci√≥n en el modelo predictivo, estas categor√≠as se redujeron a **dos clases**:

- **En Peligro (1)**: Incluye las especies clasificadas como **VU (Vulnerable)**, **EN (En Peligro)**, **CR (En Peligro Cr√≠tico)**, **EW (Extinta en la Naturaleza)**.
- **No en Peligro (0)**: Incluye las especies clasificadas como **LC (Preocupaci√≥n Menor)**, **NT (Casi Amenazada)**, **EX (Extinta)**.

Dado que estamos trabajando con especies biol√≥gicas , en el dataset poseemos variables que agrupan las especies por datos biol√≥gicos, por ello siguiendo esta agrupaci√≥n inputaremos datos en columnas con datos nulos.

```python
def imputar_por_grupo(df, col, group_cols):
    """
    Imputa valores faltantes en la columna 'col' utilizando agrupaci√≥n por 'group_cols'.
    Si es num√©rica ‚Üí usa la media por grupo.
    Si es categ√≥rica ‚Üí usa la moda por grupo.
    Si el grupo no tiene valores ‚Üí usa media/moda global.
    """

    # Determinar si la variable es num√©rica o categ√≥rica
    es_numerica = pd.api.types.is_numeric_dtype(df[col])

    if es_numerica:
        # MEDIA por grupo
        imput_values = df.groupby(group_cols)[col].transform("mean")
        global_value = df[col].mean()   # media global si falta

    else:
        # MODA por grupo
        imput_values = df.groupby(group_cols)[col].transform(
            lambda x: x.mode().iloc[0] if not x.mode().empty else None
        )
        # moda global si todo falla
        global_value = df[col].mode().iloc[0] if not df[col].mode().empty else None

    # Rellenar: primero con el valor por grupo, luego con valor global
    df[col] = df[col].fillna(imput_values)
    df[col] = df[col].fillna(global_value)

    return df
```

```python
# variables para agrupar por atributos biol√≥gicos de cada especio que no poseen nulos
group_cols = ['Group', 'Kingdom', 'Phylum', 'Class', 'Order', 'Family']
```

```python
# estas variables/ columnas poseen m√°s de 50% de valores nulos
variables_a_imputar = [
    'CRITERIA', 'FECUNDITY', 'GENERATION_LENGTH', 'DIET_BREADTH',
    'TROPHIC_LEVEL', 'DISPERSAL_ABILITY'
]

for col in variables_a_imputar:
    df = imputar_por_grupo(df, col, group_cols)
```

```python
def imputar_ambientales(df, grupo_amb, vars_ambientales):

    # Como se muestra en el gr√°fico anterior todas son num√©ricas
    numeric_cols = vars_ambientales

    # Imputar por media seg√∫n el grupo ambiental
    df[numeric_cols] = df[numeric_cols].fillna(
        df.groupby(grupo_amb, observed=False)[numeric_cols].transform("mean")
    )

    return df
```

```python
def corrige_outliers(dfx, columnas_lst):
  for columna in columnas_lst:
    q1 = dfx[columna].quantile(0.25)
    q3 = dfx[columna].quantile(0.75)
    iqr = q3 - q1
    ul = q3 + 1.5*iqr # upper_limit
    ll = q1 - 1.5*iqr # lower_limit
    dfx.loc[dfx[columna]>ul, columna] = ul
    dfx.loc[dfx[columna]<ll, columna] = ll
  return dfx
```

```python
def drop_highly_correlated(dataframe, threshold=0.94, verbose=True):
    """Elimina variables num√©ricas redundantes por alta correlaci√≥n (|œÅ|>threshold).
    Estrategia codiciosa: remueve en cada paso la variable con mayor 'redundancia';
    en caso de empate, la de menor varianza."""
    df_num = dataframe.select_dtypes(include=[np.number]).copy()
    dropped = []
    iteration = 0

    while True:
        corr = df_num.corr().abs()
        # Considerar solo tri√°ngulo superior para encontrar pares
        upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
        # Pares por encima del umbral
        pairs = [(i, j, upper.loc[i, j]) for i in upper.index for j in upper.columns if pd.notna(upper.loc[i, j]) and upper.loc[i, j] > threshold]
        if not pairs:
            break

        # Calcular 'redundancia' = cantidad de correlaciones > threshold por variable
        redundancy = (upper > threshold).sum(axis=0) + (upper > threshold).sum(axis=1)
        redundancy = redundancy.groupby(level=0).sum()  # asegura √≠ndice correcto
        # Candidatas con mayor redundancia
        max_red = redundancy.max()
        candidates = redundancy[redundancy == max_red].index.tolist()

        # Empate ‚Üí escoger la de menor varianza (m√°s "poco informativa")
        if len(candidates) > 1:
            variances = df_num[candidates].var(numeric_only=True)
            to_drop = variances.idxmin()
        else:
            to_drop = candidates[0]

        if verbose:
            print(f"[Iter {iteration}] Umbral={threshold} ‚Üí Drop: {to_drop} (redundancia={max_red}, var={df_num[to_drop].var():.4f})")

        dropped.append(to_drop)
        df_num = df_num.drop(columns=[to_drop])
        iteration += 1

    # Aplicar drop al dataframe original
    dataframe.drop(columns=[c for c in dropped if c in dataframe.columns], inplace=True)
    return dropped
```

### Evalua diferentes algoritmos de ML en crossvalidacion

No sabemos qu√© algoritmos de ML seran buenos para este problema o qu√© configuraciones usar.
Vamos a evaluar 4 algoritmos de ML diferentes:

    - Logistic Regression (LR)
    - K-Nearest Neighbors (KNN).
    - Classification and Regression Trees (CART)
    - Support Vector Machines (SVM)
    - Random Forest

```python

pipelines = []

# SMOTE + MinMax + PCA + LogReg
pipelines.append((
    'SMOTE+MinMax+PCA+LogReg',
    make_pipeline(
        SMOTE(random_state=RANDOM_STATE),
        MinMaxScaler(),
        PCA(),
        LogisticRegression(solver='liblinear', class_weight='balanced')
    )
))

# SMOTE + RF200-Gini
pipelines.append((
    'SMOTE+RF200-Gini',
    make_pipeline(
        SMOTE(random_state=RANDOM_STATE),
        RandomForestClassifier(
            n_estimators=200,
            criterion='gini',
            max_features='sqrt',
            bootstrap=True,
            min_samples_split=5,
            min_samples_leaf=2,
            random_state=42,
            n_jobs=-1,
            class_weight='balanced'
        )
    )
))

# SMOTE + Std + PCA + LogReg
pipelines.append((
    'SMOTE+Std+PCA+LogReg',
    make_pipeline(
        SMOTE(random_state=RANDOM_STATE),
        StandardScaler(),
        PCA(),
        LogisticRegression(solver='liblinear', class_weight='balanced')
    )
))

# SMOTE + MinMax + LogReg
pipelines.append((
    'SMOTE+MinMax+LogReg',
    make_pipeline(
        SMOTE(random_state=RANDOM_STATE),
        MinMaxScaler(),
        LogisticRegression(solver='liblinear', class_weight='balanced')
    )
))

# SMOTE + NNCART (Decision Tree)
pipelines.append((
    'SMOTE+NNCART',
    make_pipeline(
        SMOTE(random_state=RANDOM_STATE),
        DecisionTreeClassifier(max_depth=None, class_weight='balanced')
    )
))

# SMOTE + RF500-Gini
pipelines.append((
    'SMOTE+RF500-Gini',
    make_pipeline(
        SMOTE(random_state=RANDOM_STATE),
        RandomForestClassifier(
            n_estimators=500,
            criterion='gini',
            max_features='sqrt',
            bootstrap=True,
            min_samples_split=5,
            min_samples_leaf=2,
            random_state=RANDOM_STATE,
            n_jobs=-1,
            class_weight='balanced'
        )
    )
))

# SMOTE + RF200/MAXD-15-Gini
pipelines.append((
    'SMOTE+RF200/MAXD-15-Gini',
    make_pipeline(
        SMOTE(random_state=RANDOM_STATE),
        RandomForestClassifier(
            n_estimators=200,
            max_depth = 15,
            criterion='gini',
            max_features='sqrt',
            bootstrap=True,
            min_samples_split=10,
            min_samples_leaf=1,
            random_state=RANDOM_STATE,
            n_jobs=-1,
            class_weight='balanced'
        )
    )
))

# SMOTE + XGBoost
pipelines.append((
    'SMOTE+XGBoost',
    make_pipeline(
        SMOTE(random_state=RANDOM_STATE),
        XGBClassifier(
            random_state=RANDOM_STATE,
            n_jobs=-1,
            eval_metric='logloss'
        )
    )
))
```

```python
# Evaluamos cada pipeline de ML en estrategia de 10-fold-CV
results = []
names = []

# genera el particionamiento de 5 folds que seran usados en cada evaluacion
seed = 7
kfold = StratifiedKFold(n_splits=5, random_state=RANDOM_STATE, shuffle= True)  # especifica el particionador de datos a 5-folds CV

#  evalua cada pipeline en crosvalidacion
for pipelinename, pipeline in pipelines:
    cv_results = cross_val_score(pipeline, X_train, y_train, cv=kfold, scoring='balanced_accuracy')
    results.append(cv_results)
    names.append(pipelinename)
    print("{}: {} ({})".format(pipelinename, cv_results.mean(), cv_results.std()))
```

Los resultados demuestran que RF y XGBoost est√°n dando mejores resultados

```python
# Muestra graficamente los resultados de balanced_accuracy en crossvalidacion de los Algoritmos evaluados
fig = plt.figure(figsize=[12,6])
fig.suptitle('balanced_accuracy obtenidas en 5-fold-CV')
ax = fig.add_subplot(111)
plt.boxplot(results)
# Rotar etiquetas del eje X
ax.set_xticklabels(names, rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Optimizamos Random Forest Classifier

```python
from sklearn.model_selection import GridSearchCV
# ============================================================================
#  RANDOM FOREST
# ============================================================================

resultados={}

rf_pipeline = make_pipeline(
    SMOTE(random_state=RANDOM_STATE),
    RandomForestClassifier(
        random_state=RANDOM_STATE,
        n_jobs=-1,
        class_weight='balanced'
    )
)

param_grid_rf = {
    'randomforestclassifier__n_estimators': [100, 200],
    'randomforestclassifier__max_depth': [10, 15, None],
    'randomforestclassifier__min_samples_split': [5, 10],
    'randomforestclassifier__min_samples_leaf': [1, 2],
    'randomforestclassifier__max_features': ['sqrt', 'log2'],
    'randomforestclassifier__criterion': ['gini', 'entropy']
}

grid_rf = GridSearchCV(
    rf_pipeline,
    param_grid_rf,
    cv=kfold,
    scoring='balanced_accuracy',
    n_jobs=-1,
    verbose=1
)

print("Entrenando Random Forest con GridSearchCV...")
grid_rf.fit(X_train, y_train)

# guardamos los resultados
resultados['Random Forest'] = {
    'model': grid_rf.best_estimator_,
    'best_params': grid_rf.best_params_,
    'cv_score': grid_rf.best_score_
}

print(f"Mejor balanced_accuracy (CV): {grid_rf.best_score_:.4f}")
print(f"Mejores par√°metros: {grid_rf.best_params_}")
```

Optimizamos XGBoost
